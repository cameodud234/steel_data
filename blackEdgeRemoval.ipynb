```python
import os
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
imgPath = './boxed_images'
imagePathFile = os.path.join(imgPath, '0bae1d71e.jpg')
```

```python
img = cv2.imread(imagePathFile)
plt.imshow(img)
```

```python
import imageio
image = imageio.imread(imagePathFile)
if(len(image.shape)<3):
      print('gray')
elif len(image.shape)==3:
      print('Color(RGB)')
else:
      print('others')
```

```python
imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(imgGray)
```

```python
ret, thresh = cv2.threshold(imgGray, 150, 255, cv2.THRESH_BINARY)
```

```python
plt.imshow(thresh)
```

```python
contours, hierarchy = cv2.findContours(image=thresh, mode=cv2.RETR_TREE, method=cv2.CHAIN_APPROX_NONE)
```

```python
image_copy = imgGray.copy()
cv2.drawContours(image=image_copy, contours=contours, contourIdx=-1, color=(0, 255, 0), thickness=2, lineType=cv2.LINE_AA)

```
